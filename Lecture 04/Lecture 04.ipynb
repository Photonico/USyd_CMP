### Repulisive Force:

* Interatomic potential:
$$ \phi(r)=-\frac{B}{r^m}+\frac{A}{r^n} $$
$$ n>m $$

* Interatomic force:
$$ \vec{F}(\vec{r})=-\nabla\phi(\vec{r}) $$
$$ F(r) = -\frac{\rm d}{{\rm d}r}\phi(r) $$


```python
#### Interatonic potential

import numpy as np

A = 1
B = 2
m = 2
n = 3

step = 0.01

r = np.array(np.arange(0,10,step))

# Repulsive potential
rep = A/(r**n)
# Attractive potential
atp = -B/(r**m)
# Resulting potential
pot = rep + atp

import matplotlib; import matplotlib.pyplot as plt

params = {"text.usetex":True, "font.family":"serif", "mathtext.fontset":"cm", "axes.titlesize": 16, "axes.labelsize":14}
matplotlib.rcParams.update(params)

plt.ticklabel_format(style="sci", scilimits=(0,0))
plt.title("Interatomic potential")
plt.xlabel(r"Interatomic distance"); plt.ylabel(r"Interatomic potential")

plt.plot(r, rep, label="Repulsive potential")
plt.plot(r, atp, label="Attractive potential")
plt.plot(r, pot, label="Resulting potential")
plt.plot(r, r*0, color="gray", alpha=0.4)

plt.xlim(0, 8) 
plt.ylim(-4, 4)

plt.legend(loc="best")
plt.show()
```

```python
#### Interatonic force

import numpy as np

A = 1
B = 2
m = 2
n = 3

step = 0.01

r = np.array(np.arange(0,10,step))

# Repulsive force
ref = -n*A/(r**(n+1))
# Attractive force
atf = m*B/(r**(m+1))
# Resulting force
force = rep + atp

import matplotlib; import matplotlib.pyplot as plt

plt.ticklabel_format(style="sci", scilimits=(0,0))
plt.title("Interatomic force")
plt.xlabel(r"Interatomic distance"); plt.ylabel(r"Interatomic force")

plt.plot(r, ref, label="Repulsive force")
plt.plot(r, atf, label="Attractive force")
plt.plot(r, force, label="Resulting force")
plt.plot(r, r*0, color="gray", alpha=0.4)

plt.xlim(0, 8) 
plt.ylim(-4, 4)

plt.legend(loc="best")
plt.show()
```

### Turning Atoms in Ions

Example ${\rm NaCl}$

* Ionization energy ${\rm Na}$: $5.1{\rm eV}$

* electron affinity ${\rm Cl}$: $3.6{\rm eV}$

* net energy cost: $5.1{\rm eV}-3.6{\rm eV}=1.5{\rm eV}$ per pair

### Ionic bonding

Example ${\rm NaCl}$

* Potential energy: 
$$ E_{Coulomb} = -\frac{e^2}{4\pi\epsilon_0 a} $$

* Consider the centre ${\rm Na}$ ion, energy gain from next 6 ${\rm Cl}$ ions:
$$ E = -6\frac{e^2}{4\pi\epsilon_0 a}$$

* Energy loss from next 12  ${\rm Na}$ ions:
$$ E = +12\frac{e^2}{4\pi\epsilon_0 a\sqrt{2}}$$

* Energy gain from 8 more ${\rm Cl}$ ions:
$$ E = -8\frac{e^2}{4\pi\epsilon_0 a\sqrt{3}}$$

* The total energy:
$$ E = -\frac{e^2}{4\pi\epsilon_0 a}\times(6-\frac{12}{sqrt{2}}-\frac{8}{\sqrt{3}}) $$

* For one ion, the series converges:
$$ E \approx -1.748\frac{e^2}{4\pi\epsilon_0a} = -M_d\frac{e^2}{4\pi\epsilon_0a} $$

* $M_d$ is called the madelung constant. It is specific for a given structure.

### The covalent bonding

* Electron configuration: $2s$ and $2p$ electrons;

* Formation of four $sp^3$ hybrid orbitals as linear combination between the $s$ and three $p$ orbitals;

* Directional character of $p$ orbitals is also found in $sp^3$ orbitals.

* Cohensive energies sillilar to ionic bonding, in the $\rm eV$ range.

* Very directional bonding.

### Metallic bonding

* Outer electrons are deloaclized and act as "glue" between positively charged ion cores;

* Generally found for elements with one, two, or three valence electrons;

* Cohesive energies in the $\rm eV$ range.

### More characteristics of Metallic bonding in Simple metals

* Smaller cohesive energies than in ionic crystals;

* Larger ionic radii,   
e.g. for $\rm Na$: $3.82 {\rm A}$(metal); $1.94 {\rm A}$($\rm NaCl$);

* Bonding has no directional perference;

* Closed-packed atomic configurations are preferred: best possible overlap between the orbitals.

### Metallic bonding: Why is this so favorable?

* In Schrodinger equation:
$$ i\hbar\frac{\partial}{\partial t}\psi(\vec{r},t) = -\frac{\hbar^2}{2m}\nabla^2\psi(\vec{r},t) + V(\vec{r},t)\psi(\vec{r},t) $$
The part of kinetic energy is proportional to (negative) average curvature of wave function; "Flatter" wave function means lower energy;

* For heisenberg uncertainty principle:
$$ \Delta x\Delta p \geq \hbar/2 $$
The less localization means smaller $p$ variation, and smaller kinetic energy;




### Transition metals

* $4s$ and $3d$ have very similar energies;

* $4s$ electrons form deloaclized metallic bonds;

* $3d$ electrons form more local(covalent-like) bonds;

* Higher cohesive energies.
